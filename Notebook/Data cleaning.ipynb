```python
data['age_group']=pd.cut(
    data['customer_age'],
    bins=[17,25,35,45,55,65,100],
    labels=['18-25','26-35','36-45','46-55','56-65','65+']

)
data['age_group'].isnull().sum()
```

```python
data['risk_score'] = (
    data['velocity_flag'].astype(int) +
    data['geolocation_mismatch'].astype(int) +
    data['device_fingerprint_new'].astype(int) +
    data['refund_requested'].astype(int)
)
```

```python
df['price_category'] = np.where(
    df['price'] < df['price'].median(),
    'Low Price',
    'High Price'
)
```

```python
df['seller_group'] = np.select(
    [
        df['seller_rating'] >= 4.5,
        df['seller_rating'] >= 3.5
    ],
    [
        'High',
        'Medium'
    ],
    default='Low'
)
```

```python
df['risk_score'] = (
    df['ip_location_mismatch'].astype(int)
    + df['unusual_payment_patterns'].astype(int)
)
```
